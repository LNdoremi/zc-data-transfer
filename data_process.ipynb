## 点位表

```python
points = pd.read_excel('./excels/点位.xlsx')
point_collections = pd.read_excel('./excels/点位采集.xlsx')
towns = pd.read_excel('./excels/镇街.xlsx')
photos = pd.read_excel('./excels/照片.xlsx')
handle_records = pd.read_excel('./excels/处理记录.xlsx')
police_list = pd.read_excel('./excels/巡查员.xlsx')
volunteer_list = pd.read_excel('./excels/志愿者.xlsx')
point_assign = pd.read_excel('./excels/点位分配.xlsx')
point_archive = pd.read_excel('./excels/点位销号.xlsx')
```

```python
issues = list()

for i in range(len(points)):
    issueId = points['Id'][i]
    areaId = points['镇街Id'][i]
    ### 查找点位镇街
    for town in range(len(towns)):
        if areaId == towns['Id'][town]:
            townName = str(towns['镇街名称'][town])
    
    ### 查找点位对应的照片id list
    allPicList = list()
    picList = list()
    handlePicList = list()

    for collection in range(len(point_collections)):
        if issueId == point_collections['点位Id'][collection]:
            allPicList.append(int(point_collections['照片Id'][collection]))
    
    for picid in allPicList:
        for photo in range(len(photos)):
            if picid == photos['Id'][photo]:
                if photos['照片类型Id'][photo] == 1:
                    picList.append(photos['文件路径'][photo])
                elif photos['照片类型Id'][photo] == 2:
                    handlePicList.append(photos['文件路径'][photo])
    
    ### 查找处理记录
    for record in range(len(handle_records)):
        if issueId == handle_records['点位Id'][record]:
            handleAt = datetime.strptime(handle_records['处理时间'][record], "%m/%d/%Y %H:%M:%S").isoformat()
            handleDesc = str(handle_records['处理记录描述'][record])
            handleById = str(handle_records['处理人Id'][record])
            ### 查找处理人姓名
            for police in range(len(police_list)):
                print(police_list)
                if handleById == police_list['UserId'][police]:
                    handleByName = police_list['姓名'][police]

    ### 查找点位状态
    issueStatus = int(points['点位状态Id'][i])

    ### 查找问题类型
    issueNo = int(points['问题类型Id'][i])
    if issueNo == 1:
        issueType = '涉污水体'
    elif issueNo == 2:
        issueType = '散乱污'
    elif issueNo == 3:
        issueType = '涉嫌违建'
    elif issueNo == 4:
        issueType = '畜禽养殖'
    elif issueNo == 5:
        issueType = '涉污排水口'
    elif issueNo == 6:
        issueType = '垃圾黑点'

    create_time = datetime.strptime(points['创建时间'][i], "%m/%d/%Y %H:%M:%S").isoformat()
    ### 查找创建人姓名
    for vol in range(len(volunteer_list)):
        if points['创建人Id'][i] == volunteer_list['UserId'][vol]:
            createByName = volunteer_list['姓名'][vol]

    ### 查找分配信息
    for assign in range(len(point_assign)):
        if issueId == point_assign['点位Id'][assign]:
            assignBy = point_assign['分配人Id'][assign]
            assignTo = point_assign['负责人Id'][assign]
            assignAt = datetime.strptime(point_assign['分配时间'][assign], "%m/%d/%Y %H:%M:%S").isoformat()
    
    ### 查找销号信息
    for archive in range(len(point_archive)):
        if issueId == point_archive['点位Id'][archive]:
            archiveBy = point_archive['销号人Id'][archive]
            archiveAt = datetime.strptime(point_archive['销号时间'][archive], "%m/%d/%Y %H:%M:%S").isoformat()

    issues.append({
        "id" : int(issueId),
        "issueType": issueType,
        "status" : issueStatus,
        "riverName": str(points['河涌名'][i]),
        "createBy" : str(points['创建人Id'][i]),
        # "createByName": createByName,
        "createAt" : {
            "$date": '{}.000Z'.format(create_time)
            },
        "town" : townName,
        "areaId": int(areaId),
        "address" : str(points['位置'][i]),
        "desc" : str(points['描述'][i]),
        "latitude" : str(points['纬度'][i]),
        "longitude" : str(points['经度'][i]),
        "pics" : picList,
        "handleAt" : {
            "$date": '{}.000Z'.format(handleAt)
            },
        "handleBy" : handleById,
        "handleDesc" : handleDesc,
        "handlePicList": handlePicList,
        # "handleByName": handleByName,
        "assignBy": assignBy,
        "assignAt": {
            "$date": '{}.000Z'.format(assignAt)
            },
        "assignTo": assignTo,
        "archiveBy": archiveBy,
        "archiveAt": {
            "$date": '{}.000Z'.format(archiveAt)
            },
        "type": 'hczc'
    })

with open('./pd/issues.json', 'w', encoding='utf-8') as f :
    json.dump(issues, f, ensure_ascii=False)
```
